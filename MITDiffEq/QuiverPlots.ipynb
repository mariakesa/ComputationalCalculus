# Geometric view point on differential equations

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#inspired by http://systems-sciences.uni-graz.at/etextbook/sw3/multivariate.html
def dydx(x,y):
    return y/x

# define dimensions for grid
u = np.linspace(-10, 10,25)
v = np.linspace(-10, 10, 25)
# create grid
x, y  = np.meshgrid(u, v)
print(x.shape)

dy=dydx(x,y)


dx=x

soln=integrate.odeint(dydx,y0=[10],t=u)

plt.figure(figsize=(10, 10))
plt.quiver(x,y,dx,dy)
plt.plot(u,soln.flatten())
```
