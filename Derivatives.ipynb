```python
import numpy as np
import matplotlib.pyplot as plt
```

The derivative of a sine function is the cosine function. We use this fact to demonstrate three definitions of the derivative.

```python
derivative=np.cos(0.5)
print('Value of derivative of sine function at point x=0.5',derivative)
x_range=np.arange(start=0,stop=0.51,step=0.5/100)
print(x_range)
approximate_derivative=(np.sin(x_range)-np.sin(0.5))/(x_range-0.5)
print(approximate_derivative)
plt.plot(x_range,approximate_derivative)
```

```python
derivative=np.cos(0.5)
print('Value of derivative of sine function at point x=0.5',derivative)
h_range=np.arange(0,0.1,0.00001)
h_range=h_range[::-1]
print(h_range)
approximate_derivative=(np.sin(0.5+h_range)-np.sin(0.5))/h_range
print(approximate_derivative)
plt.plot(h_range,approximate_derivative)
```

```python
print('Value of sin function',np.sin(0.5+0.01))
print('Value of approximation',np.sin(0.5)+np.cos(0.5)*0.01)
```
