```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,101)
print(x.shape)
y=x**2
print(y.shape)
first_derivative=np.diff(y)
print(first_derivative.shape)
second_derivative=np.diff(first_derivative)
print(second_derivative.shape)
plt.plot(x,y,label='X squared')
plt.plot(x[0:-1],first_derivative,label='first derivative')
plt.plot(x[1:-1],second_derivative,label='second derivative')
plt.legend()
```

```python
plt.plot(x[0:-1],first_derivative)
plt.title('First derivative')
```

```python
print(second_derivative)
plt.plot(second_derivative)
plt.title('Second derivative')
#Notice the tiny numerical round-off
```
