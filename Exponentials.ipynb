```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The exponential function e\*\*x can be expressed as a series.

```python
def exponential_series(order):
    single_member=lambda x,n: (1./np.math.factorial(n))*(x**n)
    series_lst=[]
    x=np.arange(-10,10,0.001)
    for j in range(0,order):
        series_lst.append(single_member(x,j))
    return series_lst

order=10
series=exponential_series(order)
```

```python
for polynomial in series:
    plt.plot(np.arange(-10,10,0.001),polynomial)
plt.show()
series=np.array(series)
plt.plot(np.arange(-10,10,0.001),np.sum(series,axis=0))
```

```python
def series_with_imaginary_numbers(order):
    single_member=lambda x,n: (1./np.math.factorial(n))*((x*1j)**n)
    series_lst=[]
    x=np.arange(-10,10,0.001)
    for j in range(0,order):
        series_lst.append(single_member(x,j))
    return series_lst
im_series=series_with_imaginary_numbers(10)
im_series=np.array(im_series)
print(im_series)
```

```python
real_series_cos=im_series[0::2,:]
print(real_series_cos)
for index in range(0,real_series_cos.shape[0]):
    plt.plot(np.arange(-10,10,0.001),real_series_cos[index,:])
```

```python
im_series_sin=im_series[1::2,:]
print(im_series_sin)
for index in range(0,real_series_cos.shape[0]):
    plt.plot(np.arange(-10,10,0.001),im_series_sin[index,:].imag)
```
