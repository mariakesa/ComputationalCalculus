```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Here's an insight: derivatives can be linear functions of a state vector.

$\dot{x}=Ax$

The columns of A are the basis vectors for the dynamics.

How do we plot and understand it?

Let's see it in action for a system.



```python
A=np.random.normal(size=(2,2))
print(A)

nx, ny = (25, 25)
x = np.linspace(0, 1, nx)
y= np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)

z=[]
for one in range(nx):
    for two in range(ny):
        der_x=A@np.array([one,two])
        z.append(der_x)
        
z=np.array(z)

plt.figure(figsize=(10, 10))
plt.quiver(xv,yv,z[:,0].flatten(),z[:,1].flatten())
```
